### `loc` vs `iloc`
The tutorial doesn't make it clear the difference between `loc` and `iloc`, but the main difference is that `iloc` stands for index location, while `loc` stands for location. To make it clear, let us use an example

```python
import numpy as np
import pandas as pd

# Don't worry about this code
df = pd.DataFrame(index=[x for x in 'abcde'])
df['col1'] = [f'{x}_col1' for x in 'abcde']
df['col2'] = [f'{x}_col2' for x in 'abcde']
df
```

```python
print('position (0, 1): ', df.iloc[0, 1])  # 0 row, 1 column
print('position (4, 0): ', df.iloc[4, 0])  # 4 row, 0 column
```

```python
print('position (a, col1): ', df.loc['a', 'col1'])  # a row, col1 column
print('position (d, col2): ', df.loc['d', 'col2'])  # d row, col2, column
```

```python
print('position (a, col1): ', df.iloc['a', 'col1'])  # a row, col1 column
print('position (d, col2): ', df.iloc['d', 'col2'])  # d row, col2, column
```

```python
print('position (0, 1): ', df.loc[0, 1])  # 0 row, 1 column
print('position (4, 0): ', df.loc[4, 0])  # 4 row, 0 column
```

```python
df
```

So with `loc` you need to use the exact labels as it occurs in the DataFrame, while `iloc` requires that you use the actual index (or number) at where you want

### Something to Note when using `loc` and `iloc`

Notice the difference between `loc` and `iloc`

```python
data.head()
```

```python
data.iloc[1:3, :]
```

```python
data.loc[1:3, :]
```

This might make it clearer

```python
data.loc[:, 'Rating':'Student'].head()
```

With `data.iloc[1:3, :]` it means to start at index 1 and end RIGHT BEFORE 3, but don't include index 3.

With `data.loc[1:3, :]` it means to start and location (not index) 1 and end AT 3, so you include 3.

It is more clear with `data.loc[:, 'Rating':'Student']`. You can see that we start at `Rating` and we pick the columns until we end at `Student`, while keeping the `Student` column

This is a very slight difference, and it will probably never affect you. But if you ever want to slice the data, it is usually wise to use `iloc` over `loc`.

### When to use `loc` vs `iloc`

Anything that you can do with `loc` you can do in `iloc`. Moreover, anything you can do with `iloc` you can do with `loc`. So if that is the case, why is there a distinction? The simple answer is that some things are easier when you use `iloc` and others are easier when you use `loc`.

So the good thing that `loc` allows you to do is to use the actual names of the columns, as opposed to the index of the column. As an example, imagine I wanted the `Student` column of the dataset. With `loc` this is easy

```python
data.loc[:, 'Student'].head(2)
```

But with `iloc` you now need to remember where the location of student is. Can you remember which index it is, because I sure as hell don't. I'm not even bothered to count which index it is!

```python
data.head()
```

Also, when you are performing conditional selection, what we did right above, it is typically to use `loc`. But you can use `iloc` as well.

Now `iloc` is very useful when you want to slice the dataset, meaning when you want to take chunks of the dataset. Consider you now want the 0th column to the 3rd column of the dataset. This is very easy with splice

```python
data.iloc[:, 0:4].head()  # Remember we don't include the 4 column
```

But, if you want to use `loc` you now need to remember what the name of the 0th column and the 3rd column was, so it becomes a bit harder.

This splicing becomes much more handy when you are splicing the rows, but the idea still applies.

In summary, `loc` and `iloc` are basically the same thing. With `loc` you need to use the names, while `iloc` you need to remember the positions (or indices). There will be some situations where it is easier to remember names, there are other where positions are easier to remember. So you will simply use which ever method is easiest for you.

### Excel vs Csv

Excel and csv are basically the same thing, with the difference being that excel compresses the file as well, while csv does no compression. The result is that csv files tends to be larger than excel files are.

My choice is to use csv over excel, the main reason being that when you want to use an excel file, you need to uncompress it before it can be used. But csv can just be read with no compression, and so is generally faster to be imported into Python. 

Now for the datasets that we will work on, this will probably have no measurable impact (you'll probably only see a difference of < 1 sec). But when you work with larger datasets (~100,000 points) you will start to feel the difference (2 sec for csv to 2 minutes for excel).

### Anscombe's Quartet - aka. Statistics Lie - aka. 68.5% of all statistics are made up

So the Anscombe's dataset is quite a famous dataset in statistics. It is basically used to show why EDA is so important, and just looking at statitics can be incredibly missleading. So below I have the actualy dataset, and let us look at the statistics

```python
import pandas as pd
from scipy.stats import pearsonr
```

```python
def load_anscombe_dataset():
    """ Returns the Anscombe Dataset as a pd.Dataframe """
    
    x = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
    y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
    y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
    y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
    x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
    y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]

    datasets = {
        'I': (x, y1),
        'II': (x, y2),
        'III': (x, y3),
        'IV': (x4, y4)
    }
    
    df = pd.DataFrame()

    index_tuples = [(key, 'x') for key in datasets.keys()]
    index_tuples += [(key, 'y') for key in datasets.keys()]
    index = pd.MultiIndex.from_tuples(sorted(index_tuples))
    
    df = pd.DataFrame(columns=index, index=np.arange(len(x)))
    for key, val in datasets.items():
        df[key, 'x'] = val[0]
        df[key, 'y'] = val[1]
    return df
```

```python
dataset = load_anscombe_dataset()
dataset
```

So why is this dataset so misleading, well it is because if we look at some high level statistics of this dataset

```python
print(dataset.loc[:, pd.IndexSlice[:, 'y']].mean())

print('-'*32)

print(dataset.loc[:, pd.IndexSlice[:, 'x']].mean())
```

```python
print(dataset.loc[:, pd.IndexSlice[:, 'y']].std())

print('-'*32)

print(dataset.loc[:, pd.IndexSlice[:, 'x']].std())
```

```python
for index in ['I', 'II', 'III', 'IV']:
    correlation = pearsonr(dataset[index, 'x'], dataset[index, 'y'])[0]
    print(f'{index}: {correlation}')
```

So far we have nothing too harmful. All we have is 4 datasets where all the `x` and `y` values have the same mean, standard deviation, and the correlation between the `x` and `y` are all the same. But the trick comes when we plot the dataset.

So the statistics say a very very different story than what the actual data is saying. Now the only way to see if the statistics are misleading you, all you need to do is to plot the dataset.

### Residual Testing

Once we fit a model, the question now is - How do we know if our fitted regression is good or bad? So if you read my last week notebook then one obvious way to do this is to see how well the model performs on the test set (as opposed to the training set).


But we actually have a step before looking at the test error, and it's called residual diagnostic. So here we take our dataset $({X}, {Y})$ and build our model, let this be $\hat{f}$. As such what our model predicts is

$$\hat{Y}=\hat{f}({X})$$

Now one of the main assumptions that we make is that our dataset is an addtive error model, meaning
$$Y=f(x)+\varepsilon$$
with $\mathbb{E}[\varepsilon]=0$.

So, when we look at the residual (the difference between the true $Y$ and our prediction $\hat{Y}$) we get
$$Y-\hat{Y}=f(x)+\varepsilon - \hat{f}(X)=[f(x)-\hat{f}(x)]+\varepsilon$$

Now, we would hope that our model $\hat{f}$ is as close to the true function $f$ as possible. Meaning that we hope that 
$$\hat{f}(X)\approx f(x)\implies f(x)-\hat{f}(x)\approx 0$$

So then
$$Y-\hat{Y}=[f(x)-\hat{f}(x)]+\varepsilon\approx \varepsilon \implies Y-\hat{Y}\approx \varepsilon$$

Now remember that $\varepsilon$ is a random variable with $\mathbb{E}[\varepsilon]=0$. As such, when we look at the residuals $Y-\hat{Y}$ we should expect something to look random with about a 0 average.

So, let us look at some residuals


- On the Left : Here our residuals look quite random, and randomly distributed around 0. So this is good
- In the Middle : So while the average of the residuals seems to be about zero, one big issues is that the variance is not constant as the size of the fitted value changes. 

    To make sense of this think about predicting house prices. Then I tell you that for small prices then the error in my model is relatively small. But as prices increases then the error in my model becomes bigger and bigger. But this is quite weird, there is no reason why our model would be more accuract when the prices are small, and there is no reason why it gets less accurate as the prices increases.
    
    So eventhough there is not much of an issue with $\mathbb{E}[\varepsilon]=0$, the issues is with non-constant variance, also known as Heteroscedasity. To fix this, typically you transform the output variable using the log transformion. So instead of trying to predict $Y$, we instead try to predict $\log(Y)$ and build the model
    $$\log(Y)=\hat{f}(X)$$
- On the Right : So here hopefully everyone can see a parabolic pattern. This would indicate that there is something very fundamental about the model you are missing out on. Perhaps you are missing out on a non-linear term, or an interaction term, or perhaps you need to introduce a completely new variable to your regression function.
